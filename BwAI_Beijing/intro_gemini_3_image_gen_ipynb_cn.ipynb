### æ–‡æœ¬åˆ°å›¾åƒ

åœ¨ä¸‹é¢çš„å•å…ƒæ ¼ä¸­ï¼Œæ‚¨å°†è°ƒç”¨ `generate_content` æ–¹æ³•å¹¶ä¿®æ”¹ä»¥ä¸‹å‚æ•°ï¼š

  - `prompt`ï¼šä»…åŒ…å«æ–‡æœ¬çš„ç”¨æˆ·æ¶ˆæ¯ï¼Œæè¿°è¦ç”Ÿæˆçš„å›¾åƒã€‚
  - `config`ï¼šç”¨äºæŒ‡å®šå†…å®¹è®¾ç½®çš„é…ç½®ã€‚
    - `response_modalities`ï¼šè¦ç”Ÿæˆå›¾åƒï¼Œæ‚¨å¿…é¡»åœ¨ `response_modalities` åˆ—è¡¨ä¸­åŒ…å« `IMAGE`ã€‚è¦åŒæ—¶è·å–æ–‡æœ¬å’Œå›¾åƒï¼Œè¯·æŒ‡å®š `IMAGE` å’Œ `TEXT`ã€‚
    - `ImageConfig`ï¼šè®¾ç½® `aspect_ratio`ã€‚æœ‰æ•ˆæ¯”ä¾‹ä¸ºï¼š1:1, 3:2, 2:3, 3:4, 4:3, 4:5, 5:4, 9:16, 16:9, 21:9

æ‰€æœ‰ç”Ÿæˆçš„å›¾åƒéƒ½åŒ…å«ä¸€ä¸ª [SynthID æ°´å°](https://deepmind.google/technologies/synthid/)ï¼Œå¯ä»¥é€šè¿‡ [Vertex AI Studio](https://cloud.google.com/generative-ai-studio?hl=en) ä¸­çš„ Media Studio è¿›è¡ŒéªŒè¯ã€‚

```
prompt = """
ç”Ÿæˆä¸€ä»½å­£èŠ‚æ€§å†œäº§å“æŒ‡å—ä¿¡æ¯å›¾ã€‚åŒ…æ‹¬æœˆä»½å’Œæ¯ä¸ªå­£èŠ‚æœ‰è¶£çš„ç±»åˆ«åç§°ï¼Œä»¥åŠå†œäº§å“çš„è¯¦ç»†æ’å›¾ã€‚
"""
response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=types.GenerateContentConfig(
        response_modalities=['IMAGE', 'TEXT'],
        image_config=types.ImageConfig(
            aspect_ratio="16:9",
        ),
    ),
)

# å¦‚æœæœªç”Ÿæˆå›¾åƒï¼Œåˆ™æ£€æŸ¥é”™è¯¯
if response.candidates[0].finish_reason != types.FinishReason.STOP:
    reason = response.candidates[0].finish_reason
    raise ValueError(f"æç¤ºå†…å®¹é”™è¯¯: {reason}")

for part in response.candidates[0].content.parts:
    if part.thought:
        continue # è·³è¿‡æ˜¾ç¤ºæ€è€ƒè¿‡ç¨‹
    if part.inline_data:
        display(Image(data=part.inline_data.data, width=1000))
```

### æŸ¥çœ‹æ€è€ƒè¿‡ç¨‹

è¿™æ˜¯ä¸€ä¸ªæ€è€ƒæ¨¡å‹ï¼Œæ‚¨å¯ä»¥æŸ¥çœ‹ç”Ÿæˆå›¾åƒçš„æ€è€ƒè¿‡ç¨‹ã€‚

### åŸºäºæœç´¢ç»“æœçš„ç”Ÿæˆ

ä½¿ç”¨æ­¤æ¨¡å‹ï¼Œæ‚¨è¿˜å¯ä»¥ç”ŸæˆåŸºäº Google æœç´¢ç»“æœçš„å†…å®¹ã€‚è¯·æ³¨æ„ï¼Œè¯¥æ¨¡å‹ä»…åŸºäºæ–‡æœ¬ç»“æœï¼Œè€Œä¸åŸºäº Google æœç´¢ä¸­å¯æ‰¾åˆ°çš„å›¾åƒè¿›è¡Œç”Ÿæˆã€‚

è¦æ˜¾ç¤ºåŸºäºæœç´¢ç»“æœçš„æ•°æ®ï¼Œè¯·ä½¿ç”¨ä»¥ä¸‹å•å…ƒæ ¼ä¸­çš„è¾…åŠ©å‡½æ•°ã€‚

```
def print_grounding_data(response: types.GenerateContentResponse) -> None:
    """ä»¥ Markdown æ ¼å¼æ‰“å°å¸¦æœ‰å¼•ç”¨æºçš„ Gemini å“åº”ã€‚"""
    grounding_metadata = response.candidates[0].grounding_metadata
    lines = []

    if response.text:
        # å¼•ç”¨ç´¢å¼•ä»¥å­—èŠ‚ä¸ºå•ä½
        ENCODING = "utf-8"
        text_bytes = response.text.encode(ENCODING)
        last_byte_index = 0

        if grounding_metadata.grounding_supports:
            for support in grounding_metadata.grounding_supports:
                lines.append(
                    text_bytes[last_byte_index : support.segment.end_index].decode(ENCODING)
                )

                # ç”Ÿæˆå¹¶æ·»åŠ å¼•ç”¨è„šæ³¨ï¼ˆä¾‹å¦‚ï¼Œâ€œ[1][2]â€ï¼‰
                footnotes = "".join([f"[{i + 1}]" for i in support.grounding_chunk_indices])
                lines.append(f" {footnotes}")

                # æ›´æ–°ä¸‹ä¸€ä¸ªæ®µçš„ç´¢å¼•
                last_byte_index = support.segment.end_index

        # æ·»åŠ æœ€åä¸€ä¸ªå¼•ç”¨åå‰©ä½™çš„ä»»ä½•æ–‡æœ¬
        if last_byte_index < len(text_bytes):
            lines.append(text_bytes[last_byte_index:].decode(ENCODING))

    lines.append("\n\n----\n## å¼•ç”¨æ¥æº\n")

    if grounding_metadata.grounding_chunks:
        # æ„å»ºå¼•ç”¨æ¥æºéƒ¨åˆ†
        lines.append("### å¼•ç”¨å—\n")
        for i, chunk in enumerate(grounding_metadata.grounding_chunks, start=1):
            context = chunk.web or chunk.retrieved_context or chunk.maps
            if not context:
                continue

            uri = context.uri
            title = context.title or "æ¥æº"

            # å°† GCS URI è½¬æ¢ä¸ºå…¬å…± HTTPS URL
            if uri:
                uri = uri.replace(" ", "%20")
                if uri.startswith("gs://"):
                    uri = uri.replace(
                        "gs://", "https://storage.googleapis.com/", 1
                    )

            lines.append(f"{i}. [{title}]({uri})\n")
            if hasattr(context, "place_id") and context.place_id:
                lines.append(f"    - åœ°ç‚¹ ID: `{context.place_id}`\n\n")
            if hasattr(context, "text") and context.text:
                lines.append(f"{context.text}\n\n")

    # æ·»åŠ æœç´¢/æ£€ç´¢æŸ¥è¯¢
    if grounding_metadata.web_search_queries:
        lines.append(
            f"\n**ç½‘ç»œæœç´¢æŸ¥è¯¢:** {grounding_metadata.web_search_queries}\n"
        )
        if grounding_metadata.search_entry_point:
            lines.append(
                f"\n**æœç´¢å…¥å£ç‚¹:**\n{grounding_metadata.search_entry_point.rendered_content}\n"
            )
    elif grounding_metadata.retrieval_queries:
        lines.append(
            f"\n**æ£€ç´¢æŸ¥è¯¢:** {grounding_metadata.retrieval_queries}\n"
        )

    display(Markdown("".join(lines)))
```

æ¥ä¸‹æ¥ï¼Œæ‚¨å°†åˆ›å»ºä¸€ä¸ª Google Search å·¥å…·å¹¶å°†å…¶åŒ…å«åœ¨ä»¥ä¸‹è¯·æ±‚çš„ `tools` å‚æ•°ä¸­ã€‚

```
prompt = """
æœç´¢å¹¶å¯è§†åŒ–æ—§é‡‘å±±æœªæ¥ 5 å¤©çš„å¤©æ°”é¢„æŠ¥ï¼Œä»¥ç®€æ´ã€ç°ä»£çš„å¤©æ°”å›¾å‘ˆç°ã€‚æ·»åŠ æ¯å¤©å¯ä»¥ç©¿çš„æœè£…çš„è§†è§‰æ•ˆæœã€‚
"""
google_search = types.Tool(google_search=types.GoogleSearch())

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt,
    config=types.GenerateContentConfig(
        response_modalities=['TEXT', 'IMAGE'],
        image_config=types.ImageConfig(
            aspect_ratio="21:9",
        ),
        tools=[google_search],
    )
)

for part in response.parts:
    if part.text and part.thought:
      display(Markdown(part.text))
    elif part.inline_data:
      display(Image(data=part.inline_data.data, width=500))

print_grounding_data(response)
```

### å¤šè½®å›¾åƒç¼–è¾‘ï¼ˆèŠå¤©ï¼‰

åœ¨ä¸‹ä¸€èŠ‚ä¸­ï¼Œæ‚¨å°†ç”Ÿæˆä¸€ä¸ªèµ·å§‹å›¾åƒï¼Œå¹¶é€šè¿‡ä¸æ¨¡å‹èŠå¤©æ¥è¿­ä»£åœ°æ›´æ”¹å›¾åƒçš„æŸäº›æ–¹é¢ã€‚

```
chat = client.chats.create(
    model=MODEL_ID,
    config=types.GenerateContentConfig(
        response_modalities=['TEXT', 'IMAGE']
    )
)

message = "åˆ›å»ºä¸€å¼ é€æ˜é¦™æ°´ç“¶æ”¾åœ¨æ¢³å¦†å°ä¸Šçš„å›¾ç‰‡ã€‚"
response = chat.send_message(message)

# ä¿å­˜å›¾åƒæ•°æ®ï¼Œä»¥ä¾¿åœ¨ä¸‹ä¸€æ¬¡èŠå¤©æ¶ˆæ¯ä¸­ä¼ é€’
data = b''
for part in response.candidates[0].content.parts:
    if part.text:
        display(Markdown(part.text))
    if part.inline_data:
        data = part.inline_data.data
        display(Image(data=data, width=500))
```

ç°åœ¨ï¼Œæ‚¨å°†åœ¨ç°æœ‰èŠå¤©çš„æ¶ˆæ¯ä¸­åŒ…å«å…ˆå‰çš„å›¾åƒæ•°æ®ï¼Œä»¥åŠä¸€ä¸ªæ–°çš„æ–‡æœ¬æç¤ºï¼Œä»¥æ›´æ–°å…ˆå‰ç”Ÿæˆçš„å›¾åƒã€‚
