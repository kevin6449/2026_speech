## 概述

### Imagen 4

Vertex AI 上的 Imagen 4 为应用程序开发者带来了 Google 最先进的生成式 AI 功能。Imagen 4 是 Google 迄今为止最高质量的文本到图像模型。它能够以惊人的细节和多种风格创建图像以及渲染文本。了解更多关于 [Vertex AI 上的 Imagen](https://cloud.google.com/vertex-ai/generative-ai/docs/image/overview) 的信息。


在本教程中，您将学习如何使用适用于 Python 的 Google Gen AI SDK 与 Imagen 4、Imagen 4 Fast 和 Imagen 4 Ultra 模型交互，以生成展示以下特性的图像：

- 多样化的风格
- 多语言提示支持与增强
- 高质量照片级真实感
- 文本渲染
- 图像水印

### 定义一个用于显示图像的辅助函数

```
import typing

import IPython.display
from PIL import Image as PIL_Image
from PIL import ImageOps as PIL_ImageOps


def display_image(
    image,
    max_width: int = 700,
    max_height: int = 400,
) -> None:
    pil_image = typing.cast(PIL_Image.Image, image._pil_image)
    if pil_image.mode != "RGB":
        # 所有 Jupyter 环境都支持 RGB（例如，RGBA 尚未支持）
        pil_image = pil_image.convert("RGB")
    image_width, image_height = pil_image.size
    if max_width < image_width or max_height < image_height:
        # 调整大小以显示较小的 Notebook 图像
        pil_image = PIL_ImageOps.contain(pil_image, (max_width, max_height))
    IPython.display.display(pil_image)
```

### 加载 Imagen 4 图像生成模型

```
generation_model = "imagen-4.0-generate-001"
generation_model_fast = "imagen-4.0-fast-generate-001"
generation_model_ultra = "imagen-4.0-ultra-generate-001"
```

### Imagen 4

借助 Imagen 4，您可以生成具有更高提示依从性和文本渲染能力的高质量图像。运行以下单元格，查看使用 Imagen 4 生成图像的示例，其中包含要在图像中渲染的特定文本和标语。

您可以将 `aspect ratio` 配置为以下任意值：
- `1:1`
- `9:16`
- `16:9`
- `3:4`
- `4:3`

使用 Imagen 4 时，您可以将 `number_of_images` 设置为 1 到 4 之间的任何值。

您还可以将 `image_size` 设置为 `1K` 或 `2K`。

```
prompt = """
一面白墙上挂着两张装饰艺术风格的旅行海报。第一张海报上有文字：“NEPTUNE”，标语：“太阳系的瑰宝！”第二张海报上有文字：“JUPITER”，标语：“与巨人同行！”
"""

image = client.models.generate_images(
    model=generation_model,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        aspect_ratio="16:9",
        number_of_images=1,
        image_size="2K",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="ALLOW_ADULT",
    ),
)

display_image(image.generated_images[0].image)
```

### Imagen 4 Fast

除了 Imagen 4，您还可以选择使用 Imagen 4 Fast。这两个模型选项让您可以根据使用场景选择优化质量或延迟。

**Imagen 4：** 生成具有自然光照和更高照片级真实感的高质量图像。

**Imagen 4 Fast：** 适用于创建亮度更高、对比度更强的图像。

```
import matplotlib.pyplot as plt

prompt = """
一张纽约天际线日落的图片
"""

# Imagen 4 图像生成
image = client.models.generate_images(
    model=generation_model,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="3:4",
        image_size="2K",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="ALLOW_ADULT",
    ),
)

# Imagen 4 Fast 图像生成
fast_image = client.models.generate_images(
    model=generation_model_fast,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="3:4",
        image_size="2K",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="ALLOW_ADULT",
    ),
)

# 显示生成的图像
fig, axis = plt.subplots(1, 2, figsize=(12, 6))
axis[0].imshow(image.generated_images[0].image._pil_image)
axis[0].set_title("Imagen 4")
axis[1].imshow(fast_image.generated_images[0].image._pil_image)
axis[1].set_title("Imagen 4 Fast")
for ax in axis:
    ax.axis("off")
plt.show()
```

### 使用 Imagen 4 Ultra 生成高质量图像

除了 Imagen 4 和 Imagen 4 Fast，您还可以选择使用 Imagen 4 Ultra。该模型将返回质量极高的图像，但会以更高的延迟为代价。

在生成人物图像时，您还可以相应地设置 `safety_filter_level` 和 `person_generation` 参数：
- `person_generation`（人物生成）
  - `DONT_ALLOW`
  - `ALLOW_ADULT`
  - `ALLOW_ALL`
- `safety_filter_level`（安全过滤级别）
  - `BLOCK_LOW_AND_ABOVE`
  - `BLOCK_MEDIUM_AND_ABOVE`
  - `BLOCK_ONLY_HIGH`
  - `BLOCK_NONE`

```
prompt = """
照片级真实感的夜景：从寒冷的室外街道望向一家灯火通明、经典的 1960 年代美式餐厅。整个视野都透过一块布满雨痕的大玻璃窗。外面霓虹灯“DINER”招牌在湿漉漉的人行道和窗户本身投射出多彩的倒影。情绪化、怀旧、浅景深强调窗户表面。
"""

image = client.models.generate_images(
    model=generation_model_ultra,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="1:1",
        image_size="2K",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="ALLOW_ADULT",
    ),
)

display_image(image.generated_images[0].image)
```

### 多语言和提示重写器支持

Imagen 4 还能够处理多种语言的文本提示。以下是支持的文本提示语言输入值：
- 英语
- 中文（简体）
- 中文（繁体）
- 印地语
- 日语
- 韩语
- 葡萄牙语
- 西班牙语
- 德语
- 法语

`imagen-4.0-generate-001` 模型提供增强您提供的提示的选项。要使用此功能，请将 `enhance_prompt` 设置为 True。系统将从您的原始提示创建一个新的详细提示，以帮助生成更高质量的图像。

```
prompt = """
Una pintura al óleo impresionista de una taza de café sobre una mesa en una cocina, con las palabras 'buenos días' escritas en una fuente caprichosa en la taza.
"""

image = client.models.generate_images(
    model=generation_model,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="1:1",
        image_size="2K",
        enhance_prompt=True,
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="ALLOW_ADULT",
    ),
)

# 打印增强后的提示
print(image.generated_images[0].enhanced_prompt)
display_image(image.generated_images[0].image)
```

### 更好的文本渲染

Imagen 4 在图像中准确渲染单词、标题和短语方面表现出色。这对于漫画、徽标生成、流程图、传单和海报特别有用。

```
prompt = """
一幅漫画分格。一只可爱的灰猫正在和一只斗牛犬说话。这只猫看起来有点厌恶。猫在对话气泡中说：“你似乎真的很喜欢出去玩。真有趣。”狗耸了耸肩作为回应。清晰的插图，线条和阴影充满自信。
"""

image = client.models.generate_images(
    model=generation_model,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="4:3",
        image_size="2K",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="ALLOW_ADULT",
    ),
)

display_image(image.generated_images[0].image)
```

### 添加图像水印

默认情况下，Imagen 4 图像会添加数字水印或 [SynthID](https://deepmind.google/technologies/synthid/)。如果您想明确将水印设置为 True，可以使用 `add_watermark` 参数。您还可以通过 Vertex AI Studio [验证带水印的图像](https://cloud.google.com/vertex-ai/generative-ai/docs/image/generate-images#watermark)。

```
prompt = """
为时代浪漫电影《猩红之线》设计一张优雅而富有象征意义的电影海报。一个特写镜头，艺术地展现两只手，几乎触碰，一根鲜艳的猩红丝线在它们指间缠绕，巧妙地在丝线的缠绕和流动中形成标题“猩红之线”。标题的字体应感觉像流畅的手工刺绣。背景是一个柔焦、去饱和的花园。
"""

image = client.models.generate_images(
    model=generation_model,
    prompt=prompt,
    config=types.GenerateImagesConfig(
        number_of_images=1,
        aspect_ratio="1:1",
        image_size="2K",
        safety_filter_level="BLOCK_MEDIUM_AND_ABOVE",
        person_generation="ALLOW_ADULT",
        add_watermark=True,
    ),
)

display_image(image.generated_images[0].image)
```
